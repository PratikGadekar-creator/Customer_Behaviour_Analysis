```python
#create a column age group

labels=["Young_adult","Adult","Middle_aged","Senior"]
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
import sqlite3

conn = sqlite3.connect('customer_behaviour')   # in-memory database
```

```python
df.to_sql('customers', conn, index=False, if_exists='replace')
```

```python
pd.read_sql("SELECT * FROM customers", conn)
```

```python
#Q1) what is the total revenue generated by male vs female customers

query= '''
    SELECT gender, SUM(purchase_amount) as revenue
    from customers
    GROUP BY gender;
    '''
pd.read_sql(query,conn)
```

```python
#q2) which customer used a discount but still spent more than the average purchase amount .

query='''
    SELECT customer_id,purchase_amount
    FROM customers
    WHERE discount_applied='Yes' and purchase_amount >=(select AVG(purchase_amount) from customers)
'''
pd.read_sql(query,conn)
```

```python
#Q3) which are the top 5 products with highest average ratings
query='''
    SELECT item_purchased, ROUND(AVG(review_rating),2) as "Average_Product_rating"
    FROM customers
    GROUP BY item_purchased
    ORDER BY "Average_Product_rating" DESC
    LIMIT 5
'''
pd.read_sql(query,conn)
```

```python
#compare the average purchase amount between standerd and express shipping

query='''
    SELECT shipping_type, ROUND(AVG(purchase_amount),2)
    FROM customers
    WHERE shipping_type in ('Standard','Express')
    GROUP BY shipping_type
'''
pd.read_sql(query,conn)
```

```python
#Q5)do subscribed customers spend more? comapre spend and
#total revenue between subscribers and non-subscribers

query='''
    SELECT subscription_status, ROUND(AVG(purchase_amount),2) as "Average_spend",
    SUM(purchase_amount) as "Total_revenue"
    FROM customers
    GROUP BY subscription_status
    ORDER BY "Average_spend", "Total_revenue" DESC
'''
pd.read_sql(query,conn)
```

```python
#Q6)which 5 products have the highest percentage of purchase with discount applied?

query = '''
SELECT item_purchased,
       ROUND(
           SUM(CASE WHEN discount_applied = 'Yes' THEN 1 ELSE 0 END) * 100.0
           / COUNT(*),
           2
       ) AS discount_rate
FROM customers
GROUP BY item_purchased
ORDER BY discount_rate DESC
LIMIT 5
'''
pd.read_sql(query, conn)

```

```python
#Q7)segment customers into new ,returning and loyal based on their total number of purchases
# and show the count of each segment

query = '''
with customer_type as(
  select customer_id,previous_purchases,
  case
    when previous_purchases = 1 then 'New'
    when previous_purchases between 2 and 10 then 'Returning'
    else 'Loyal'
    end as customer_segment
    from customers

)

select customer_segment, count(*) as "Number_of_customers"
from customer_type
group by customer_segment
'''
pd.read_sql(query,conn)
```

```python
query='''
with item_counts as (
  select category,
  item_purchased,
  count(customer_id) as "Total_orders",
  row_number() over (partition by category order by count(customer_id) desc) as item_rank
  from customers
  group by category,item_purchased
)
  select item_purchased,category,item_purchased,total_orders
  from item_counts
  where item_rank<=3
'''

pd.read_sql(query,conn)
```

```python
#Q9)  ARE CUSOMERS ARE REPEAT BUYERS(MORE THEN 5 PURCHASES ) ALSO LIKELY TO SUBSCRIBE?

query='''
select subscription_status,
count (customer_id) as repeat_buyers
from customers
where previous_purchases>5
group by subscription_status
'''
pd.read_sql(query,conn)
```

```python
query='''

select age_group,
sum(purchase_amount) as total_revenue
from customers
group by age_group
order by total_revenue desc
'''
pd.read_sql(query,conn)
```
